# **visualizing**

```python
import matplotlib.pyplot as plt
plt.scatter(df['Rest_Energy'], df['Time'])
plt.xlabel('Time')
plt.ylabel('Rest_Energy')
plt.title('the relation between the rest energy and time')
plt.show()
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(13, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df_ = df.groupby([df['TTL'], 'Type'])['Rest_Energy'].count().reset_index()
df_.columns = ['TTL', 'type', 'rest energy']
plt.figure(figsize=(15, 6))
sns.lineplot(data=df_, x='TTL', y='rest energy', hue='type')
plt.xlabel('TTL')
plt.ylabel('rest energy')
plt.title('the behavior of attacks types among the rest energies and TTL')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))

g = sns.FacetGrid(df, col='Event', col_wrap=3, height=4, aspect=1.2)

g.map(sns.countplot, 'Type')

g.set_axis_labels('Type', 'Count')
g.set_titles('Type: {col_name}')
g.fig.suptitle('Counts of Type of Attacks in Every Type of Event', y=1.02)

g.add_legend(title='Type of Attack', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()

```

```python
selected_features=['Rest_Energy','Time']
df1_mean = df.groupby('Type')[selected_features].mean().reset_index()
plt.figure(figsize=(6, 4))
for feature in selected_features:
  sns.barplot(data=df1_mean, x='Type', y=feature)
  plt.title('Mean Comparison of '+str(feature)+' on each type')
  plt.xlabel('Type`')
  plt.ylabel('Mean Value')
  plt.show()
```
